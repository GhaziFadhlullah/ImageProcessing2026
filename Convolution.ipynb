```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# 1. Membaca gambar
img = Image.open("kopi.jpeg").convert("L")
image = np.array(img)
```

```python
# 2. Menampilkan gambar asli
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# 3. Membuat kernel rataan 5x5
kernel = np.ones((5,5)) / 25
```

```python
# 4. Menentukan ukuran gambar
rows, cols = image.shape

# membuat matriks output
output = np.zeros((rows, cols))

# padding ukuran kernel
pad = 2
```

```python
# 5. Proses konvolusi manual
for i in range(pad, rows-pad):
    for j in range(pad, cols-pad):

        total = 0

        for m in range(-pad, pad+1):
            for n in range(-pad, pad+1):

                pixel = image[i+m, j+n]
                weight = kernel[m+pad, n+pad]

                total += pixel * weight

        output[i,j] = total
```

```python
# 6. Konversi hasil ke integer
output = output.astype(np.uint8)
```

```python
# 7. Menampilkan hasil konvolusi
plt.title("Hasil Konvolusi Kernel Rataan 5x5")
plt.imshow(output, cmap='gray')
plt.axis('off')
plt.show()
```
