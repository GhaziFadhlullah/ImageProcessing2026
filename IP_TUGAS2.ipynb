```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('kopi.jpeg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print("Ukuran citra:", img_gray.shape)
```

Penjelasan:

Citra diubah ke grayscale agar matriks menjadi 2D.

Ini mempermudah operasi aritmetika piksel.

**Penjumlahan Skalar (Mencerahkan Citra)**

Misalnya skalar = 40

```python
scalar = 40

img_tambah = img_gray + scalar

# Membatasi nilai piksel 0–255
img_tambah = np.clip(img_tambah, 0, 255).astype(np.uint8)
```

Tujuan:

Meningkatkan intensitas piksel agar citra terlihat lebih terang.

**Pengurangan Skalar (Menggelapkan Citra)**

```python

scalar = 40

img_kurang = img_gray - scalar
img_kurang = np.clip(img_kurang, 0, 255).astype(np.uint8)
```

Tujuan:

Menurunkan intensitas piksel agar citra terlihat lebih gelap.

**VISUALISASI HASIL**

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Citra Asli")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img_tambah, cmap='gray')
plt.title("Tambah Skalar")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_kurang, cmap='gray')
plt.title("Kurang Skalar")
plt.axis('off')

plt.show()
```
